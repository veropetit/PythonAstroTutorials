## Loading up some data

```python
## fits files
from astropy.io import fits
url = 'https://github.com/veropetit/PythonAstroTutorials/blob/main/AstroDataFrames-data/ni3627010101.pha.fits?raw=true'
fits_hdul = fits.open(url)
fits_hdul.info

## h5 files
import h5py
url = 'https://raw.githubusercontent.com/veropetit/PythonAstroTutorials/main/AstroDataFrames-data/fl.4.h5'
wget.download(url)

h5_f = h5py.File('fl.4.h5', 'r')

```

## Getting information about the structure of the data frame

```python
## fits
fits_hdul.info()
```

```python
## h5
for key in h5_f.keys():
   print(key)
   for attr in h5_f[key].attrs:
    print('** ', attr )
## CHECK THE VISIT ITEM FuNCTION
```

```python
## closing up for the fits file:
fits_hdul.close()

## closing up for the h5 file
h5_f.close()
```
